```python
from transformers import pipeline
```

```python
summarizer = pipeline("summarization", model="sshleifer/distilbart-cnn-6-6")

text = """
The Orbiter Discovery mission was a human spaceflight mission in NASA's Space Shuttle program.
The mission was launched from Kennedy Space Center, Florida, on 24 April 1990.
It carried the Hubble Space Telescope into orbit.
The mission was the 35th flight of the Space Shuttle Discovery and the tenth mission of the Space Shuttle program.
The mission was a success, and the Hubble Space Telescope was deployed into orbit.
"""

summary = summarizer(text, max_length=50, min_length=10)

print(summary[0]['summary_text'])
```
